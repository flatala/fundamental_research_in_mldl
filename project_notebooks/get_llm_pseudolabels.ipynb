```python
"""
Traverse datasets/UCF-101/, map every video to a Places365 scene label,
and save the result to ucf101_scene_labels.csv
"""
import csv
from pathlib import Path

# ------------------------------------------------------------------
# 1. UCF-101 action  ➜  Places365 scene label
# ------------------------------------------------------------------
UCF_TO_PLACES365 = {
    "YoYo": "street",
    "WritingOnBoard": "classroom",
    "WallPushups": "gymnasium/indoor",
    "WalkingWithDog": "park",
    "VolleyballSpiking": "gymnasium/indoor",
    "UnevenBars": "gymnasium/indoor",
    "Typing": "office",
    "TrampolineJumping": "gymnasium/indoor",
    "ThrowDiscus": "athletic_field/outdoor",
    "TennisSwing": "athletic_field/outdoor",
    "TaiChi": "park",
    "TableTennisShot": "recreation_room",
    "Swing": "playground",
    "Surfing": "ocean",
    "SumoWrestling": "arena/performance",
    "StillRings": "gymnasium/indoor",
    "SoccerPenalty": "stadium/soccer",
    "SoccerJuggling": "soccer_field",
    "SkyDiving": "sky",
    "Skijet": "ocean",
    "Skiing": "ski_slope",
    "SkateBoarding": "street",
    "Shotput": "athletic_field/outdoor",
    "ShavingBeard": "bathroom",
    "SalsaSpin": "discotheque",
    "Rowing": "lake/natural",
    "RopeClimbing": "gymnasium/indoor",
    "RockClimbingIndoor": "gymnasium/indoor",
    "Rafting": "lake/natural",
    "PushUps": "gymnasium/indoor",
    "Punch": "martial_arts_gym",
    "PullUps": "gymnasium/indoor",
    "PommelHorse": "gymnasium/indoor",
    "PoleVault": "athletic_field/outdoor",
    "PlayingViolin": "music_studio",
    "PlayingTabla": "music_studio",
    "PlayingSitar": "music_studio",
    "PlayingPiano": "music_studio",
    "PlayingGuitar": "music_studio",
    "PlayingFlute": "music_studio",
    "PlayingDhol": "music_studio",
    "PlayingDaf": "music_studio",
    "PlayingCello": "music_studio",
    "PizzaTossing": "pizzeria",
    "ParallelBars": "gymnasium/indoor",
    "Nunchucks": "martial_arts_gym",
    "MoppingFloor": "corridor",
    "Mixing": "kitchen",
    "MilitaryParade": "street",
    "Lunges": "gymnasium/indoor",
    "LongJump": "athletic_field/outdoor",
    "Knitting": "living_room",
    "Kayaking": "lake/natural",
    "JumpRope": "gymnasium/indoor",
    "JumpingJack": "gymnasium/indoor",
    "JugglingBalls": "park",
    "JavelinThrow": "athletic_field/outdoor",
    "IceDancing": "ice_skating_rink/indoor",
    "HulaHoop": "gymnasium/indoor",
    "HorseRiding": "pasture",
    "HorseRace": "racecourse",
    "HighJump": "athletic_field/outdoor",
    "HeadMassage": "beauty_salon",
    "HandstandWalking": "gymnasium/indoor",
    "HandstandPushups": "gymnasium/indoor",
    "HammerThrow": "athletic_field/outdoor",
    "Hammering": "construction_site",
    "Haircut": "beauty_salon",
    "GolfSwing": "golf_course",
    "FrontCrawl": "swimming_pool/indoor",
    "FrisbeeCatch": "park",
    "FloorGymnastics": "gymnasium/indoor",
    "FieldHockeyPenalty": "athletic_field/outdoor",
    "Fencing": "gymnasium/indoor",
    "Drumming": "music_studio",
    "Diving": "swimming_pool/indoor",
    "CuttingInKitchen": "kitchen",
    "CricketShot": "athletic_field/outdoor",
    "CricketBowling": "athletic_field/outdoor",
    "CliffDiving": "cliff",
    "CleanAndJerk": "gymnasium/indoor",
    "BrushingTeeth": "bathroom",
    "BreastStroke": "swimming_pool/indoor",
    "BoxingSpeedBag": "boxing_ring",
    "BoxingPunchingBag": "boxing_ring",
    "Bowling": "bowling_alley",
    "BodyWeightSquats": "gymnasium/indoor",
    "BlowingCandles": "dining_room",
    "BlowDryHair": "beauty_salon",
    "Billiards": "recreation_room",
    "Biking": "highway",
    "BenchPress": "gymnasium/indoor",
    "BasketballDunk": "basketball_court/indoor",
    "Basketball": "basketball_court/indoor",
    "BaseballPitch": "baseball_field",
    "BandMarching": "street",
    "BalanceBeam": "gymnasium/indoor",
    "BabyCrawling": "living_room",
    "Archery": "athletic_field/outdoor",
    "ApplyLipstick": "bathroom",
    "ApplyEyeMakeup": "bathroom",
}

# ------------------------------------------------------------------
# 2.  Basic configuration
# ------------------------------------------------------------------
ROOT_DIR = Path("datasets/UCF-101")
OUT_CSV  = Path("datasets/scene_classification_results_llm.csv")
VIDEO_EXT = {".avi", ".mp4", ".mkv", ".mov"}

# ------------------------------------------------------------------
# 3.  Gather rows
# ------------------------------------------------------------------
rows = []

for action_dir in ROOT_DIR.iterdir():
    if not action_dir.is_dir():
        continue

    action = action_dir.name
    scene_label = UCF_TO_PLACES365.get(action)

    if scene_label is None:            # skip any unknown action
        print(f"[warn] '{action}' not in dict – skipped.")
        continue

    for video in action_dir.rglob("*"):
        if video.suffix.lower() not in VIDEO_EXT:
            continue

        # keep the path relative to ROOT_DIR but DROP the extension
        rel_path = video.relative_to(ROOT_DIR)      # e.g. YoYo/v_YoYo_g01_c01.avi
        rel_no_ext = rel_path.with_suffix("")       # → YoYo/v_YoYo_g01_c01
        rows.append((str(rel_no_ext), scene_label))

# ---- flip the order so the last video ends up first in the CSV ----
rows.reverse()

# ------------------------------------------------------------------
# 4.  Write CSV
# ------------------------------------------------------------------
OUT_CSV.parent.mkdir(parents=True, exist_ok=True)
with OUT_CSV.open("w", newline="") as fh:
    writer = csv.writer(fh)
    writer.writerow(["video_id", "scene_label"])
    writer.writerows(rows)

print(f"✓  Wrote {len(rows):,} rows → {OUT_CSV.resolve()}")
```
